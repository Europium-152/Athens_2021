```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_name = "e-lab_data.csv"

df = pd.read_csv(file_name, encoding='unicode_escape', decimal=',')
df.head()
```

```python
time = np.arange(df.values.shape[0]) * 0.01
voltage = df.iloc[:, 4].values
current = -df.iloc[:, 6].values
pressure = df.iloc[:, 8].values
```

```python
figure, (axis_1, axis_2, axis_3) = plt.subplots(3, 1)
axis_1.plot(time, pressure)
axis_2.plot(time, voltage)
axis_3.plot(time, current)
```
